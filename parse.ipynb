####  For simplicity we assume that in  all URLs all the  products at www.**/products/  endpoints lie at <a href=collections/  . this is for now but in future we should find more general approach 

```python
u1 = "https://vastinterior.com.au/products/"
u2 = "https://www.hudsonfurniture.com.au/products/"
u3 = "https://premiumpatio.com.au/products/"
u4 = "https://www.wellhunghammocks.com/products/"
u5 = "https://floydhome.com/products/"
u6 = "https://www.mybudgetfurniture.com/products/"
```

```python

webpage = requests.get(u5,headers=HEADERS)
soup = BeautifulSoup(webpage.content, "lxml")

#render = soup.find_all("a", href= re.compile(r'/collections/.*'))
#render = soup.find_all("span")
render = soup.select("a[href]")

render
```

```python
prods = []
for link_element in render:
   url = link_element['href']
   if "/collections/" in url:
      prods.append(url.split('/')[2])
```
